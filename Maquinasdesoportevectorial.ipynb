```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
```

```python
# Cargar el conjunto de datos de ejemplos Iris 
iris = datasets.load_iris()
X = iris.data[:, :2]
Y = iris.target
```

```python
#Crear una instancia del clasificador SVM y ajustarlo a los datos
C =  1.0 # Parametro de regularizacion
svc = svm.SVC(kernel='linear', C=C).fit(X,Y)
```

```python
# Crear una malla para grafica
h = 0.02 # Tama√±o de paso en la malla
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 0].min() - 1, X[:, 0].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
```

```python
#Realizar la clasificacion en la malla
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
#Graficar los resultados
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Paired)
plt.xlabel('Caracteristica 1')
plt.ylabel('Caracteristica 2')
plt.title('Maquinas de Soporte Vectorial')
plt.show()
```
