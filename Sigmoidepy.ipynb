```python
import numpy as np #maths library
import matplotlib.pyplot as plt # visualizacion

#Funcion sigmodeo

def sigmoid(x):
    return 1/(1+np.exp(-x))

def relu(x):
    return np.maximum(0,x)

def tanh(x):
    return np.tanh(x)

#Visualizacion 
x_values = np.linspace(-5,5,100)
print(x_values)

```

```python
plt.figure(figsize=(12,8))

plt.subplot(2,3,1) #2 filas 3 columnas, lugar o posicion 1
plt.plot(x_values,sigmoid(x_values),label='Sigmoide')
plt.title('Sigmoide')

```

```python
plt.subplot(2,3,2) #dosfilas 3 columnas lugar 2
plt.plot(x_values,relu(x_values),label='Relu')
plt.title('Relu')
```

```python
plt.subplot(2,3,3)
plt.plot(x_values,tanh(x_values),label='Tanh')
plt.title('Tanh')
```
